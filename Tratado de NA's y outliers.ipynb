```python
# Importamos las librer√≠as pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga del .csv
data=pd.read_csv('Ventas_totales.csv')
data
```

```python
valores_nulos=data.isnull().sum()
valores_nulos
```

PROCEDIMIENTO PARA ELIMINAR VALORES NULOS EN DATAFRAME

```python
#Reemplazamos NA's del dataframe con bfill
data1=data.fillna(method="bfill")
data1
```

```python
valores_nulos1=data1.isnull().sum()
valores_nulos1
```

```python
#Reemplazar los NA's con "mean"
data2=data.copy()
data2=data2.fillna(round(data.mean(),1)) #1 columnas, 0 filas
data2
```

```python
valores_nulos2=data2.isnull().sum()
valores_nulos2
```

```python
#Diagramas de caja y bigotes de cada columna del dataframe
fig = plt.figure(figsize=(20,12))
data2.plot(kind='box', vert=False)
plt.title("Valores at√≠picos del Dataset")
plt.show() #Mostramos el diagrama
```

ELIIMINAR OUTLIERS DEL DATAFRAME CON DESV. EST.

```python
#M√©todo aplicando desv est. Encuentro los valores extremos.
y=data2
LS = y.mean() + 3*y.std()
LI = y.mean() - 3*y.std()
print("Limite superior ",LS)
print("Limite inferior ",LI)
```

```python
# Encontrar Outliers dentro del Dataframe
# Los NaN son valores normales, los num√©ricos son outliers
outliers=data2[(y>LS)|(y<LI)]
outliers
```

```python
#Obtenemos datos sin outliers del Dataframe
#Convertir los outliers en NA's
data3=data2[(y<=LS)&(y>=LI)]
data3
```

```python
#Elimino las columnas innecesarias
data4=data3.drop(["indice_tiempo"],axis=1)
data4
```

```python
#Reemplazo los at√≠picos (actualmente los NA's) con mean
#Copia del dataframe
data_clean=data4.copy()
data_clean=data_clean.fillna(round(data4.mean(),1))
data_clean
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos3=data_clean.isnull().sum()
valores_nulos3
```

PROCEDIMIENTO "CUARTILES" PARA ELIMINAR OUTLIERS EN DATAFRAME

```python
#M√©todo aplicando cartiles. Encuentro cuartiles 0.25 y 0.75
y=data2

p25=y.quantile(0.25) #Q1
p75=y.quantile(0.75) #Q3
iqr=p75-p25

LS_iqr=p75 + 1.5*iqr
LI_iqr=p25 - 1.5*iqr
print("L√≠mite superior permitido",LS_iqr)
print("L√≠mite inferior permitido",LI_iqr)
```

```python
#Encontramos Outliers del Dataframe
outliers_iqr= data2[(y>LS_iqr)|(y<LI_iqr)]
outliers_iqr
```

```python
#Obtenemos los datos limpios con los outliers como NA's
data3_iqr = data2[(y<=LS_iqr)&(y>=LI_iqr)]
data3_iqr
```

```python
#Eliminamos las columnas innecesarias
data4_iqr=data3_iqr.drop(['indice_tiempo'],axis=1)
data4_iqr
```

```python
#Reemplazamos los valores at√≠picos (ahora NA's) del dataframe con "mean"
#Realizamos una copia del dataframe
data5_iqr=data4_iqr.copy()
data5_iqr=data5_iqr.fillna(round(data4_iqr.mean(),1))
data5_iqr
```

```python
#Unimos la columna que hab√≠amos eliminado

indice_tiempo=data2["indice_tiempo"]
Datos_limpios = pd.concat([indice_tiempo,data5_iqr], axis=1)
Datos_limpios
```

```python
#Corroboramos valores nulos del dataframe LIMPIO

valores_nulos_DL=Datos_limpios.isnull().sum()
valores_nulos_DL
```
